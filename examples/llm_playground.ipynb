# implicit shadow model

```python
from transformers import AutoModelForCausalLM

from shadow_peft import ShadowConfig, get_shadow_model

model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen3-0.6B")

shadow_config = ShadowConfig(
    num_shadow_layers=1,  # used for implicit shadow model init
    shadow_intermediate_size=256,
    injection_hidden_size=16,
    gate_hidden_size=10,
    alpha=0.1,  # for classification, please set it > 1, for generation, please set it < 1
    dropout=0.1,
)

model = get_shadow_model(model, shadow_config)  # wraps base model in-place
model.print_trainable_parameters()
model
```

```python
shadow_model = model.export_shadow()
```

# explicit shadow model

```python
from transformers import AutoModelForCausalLM

from shadow_peft import ShadowConfig, get_shadow_model

model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen3-0.6B")
shadow_model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen3-0.6B")

shadow_config = ShadowConfig(
    injection_hidden_size=16,
    gate_hidden_size=10,
    alpha=0.1,  # for classification, please set it > 1, for generation, please set it < 1
    dropout=0.1,
)

model = get_shadow_model(model, shadow_config, shadow_model=shadow_model)  # wraps base model in-place
model.print_trainable_parameters()
```

```python
model
```
